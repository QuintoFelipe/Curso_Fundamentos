## Solucion Ejercicio 1 

```python
import pandas as pd 
```

Creamos un DataFrame a partir del diccionario dado y usaremos las etiquetas especificadas como el indice    

```python
data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(data, index=labels)
print("DataFrame Original:\n", df)
```

### 1. Seleccione solo las columnas *'name'* y *'score'*.
Para seleccionar columnas específicas, hacemos lo siguiente:

```python
selected_columns = df[['name', 'score']]
print("Columnas seleccionadas 'name' y 'score':\n", selected_columns)
```

### 2. Filtre las filas donde el número de intentos (`attemps`) sea mayor que 2.

Podemos filtrar filas según una condición:

```python
filtered_attempts = df[df['attempts'] > 2]
print("Filas donde attempts son mayores a 2:\n", filtered_attempts)
```

### 3. Filtre las filas donde el *score* se encuentre entre 15 y 20 (incluidos)

Para filtrar basado en el puntaje, usamos:

```python
filtered_score = df[(df['score'] >= 15) & (df['score'] <= 20)]
print("Filas donde el score está entre 15 y 20:\n", filtered_score)
```

### 4. Cambie el *score* en la fila 'd' a 11.5

Podemos modificar el DataFrame especificando el índice:

```python
df.at['d', 'score'] = 11.5
print("DataFrame después de cambiar el score en la fila 'd' a 11.5:\n", df)
```

### 5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

Podemos agregar una nueva fila utilizando el método `loc`:

```python
df.loc['k'] = ['Nuevo Nombre', 18, 2, 'yes']
print("DataFrame después de agregar la nueva fila 'k':\n", df)
```

### 6. Borre la fila 'k' y entregue nuevamente el DataFrame original.

Para eliminar una fila, usamos el método `drop`:



```python
df = df.drop('k')
print("DataFrame después de eliminar la fila 'k':\n", df)
```

### 7. Ordene el DataFrame por nombre en orden ascendente.

Podemos ordenar el DataFrame utilizando el método `sort_values`:

```python
df_sorted = df.sort_values(by='name')
print("DataFrame ordenado por name:\n", df_sorted)
```

### 8. Reemplace los valores de la columna *qualify* que contiene los valores *yes* y *no* por los valores booleanos True y False respectivamente.

Finalmente, reemplazamos los valores en la columna qualify:

```python
df['qualify'] = df['qualify'].map({'yes': True, 'no': False})
print("DataFrame después de reemplazar 'yes' y 'no' en la columna qualify con True y False:\n", df)
```

### 1. Importarlo como nombre athletes

```python
import pandas as pd


# O si quieres cargarlo directamente desde el enlace:
file_path =  'athletes.csv'
athletes = pd.read_csv(file_path )

print('DataFrame athletes importado:')

athletes.head()
```

### 2.Filtrar el DataFrame para Quedarse Solo con los Atletas que Hayan Ganado al Menos una Medalla

Vamos a sumar las columnas de medallas (gold, silver, bronze) para determinar qué atletas han ganado al menos una medalla.

```python
athletes['total_medals'] = athletes['gold'] + athletes['silver'] + athletes['bronze']
athletes_with_medals = athletes[athletes['total_medals'] > 0]
print("Atletas que han ganado al menos una medalla:")
athletes_with_medals.head()
```

### 3. Calcule la masa muscular por fila con una función 

Nos podemos dar cuenta que en las columnas de peso y altura faltan datos

```python
import matplotlib.pyplot as plt

# Plotting the distributions
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Distribution for height
axes[0].hist(athletes['height'].dropna(), bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Height Distribution')
axes[0].set_xlabel('Height (cm)')
axes[0].set_ylabel('Frequency')

# Distribution for weight
axes[1].hist(athletes['weight'].dropna(), bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title('Weight Distribution')
axes[1].set_xlabel('Weight (kg)')
axes[1].set_ylabel('Frequency')
```

Usamos imputacion mediante la media de los datos para el peso y la altura

```python
# Impute missing values in 'Weight' and 'Height' with the column mean
athletes['weight'].fillna(athletes['weight'].mean(), inplace=True)
athletes['height'].fillna(athletes['height'].mean(), inplace=True)
```

Observamos la nueva distribución 

```python
# Plotting the distributions
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Distribution for height
axes[0].hist(athletes['height'].dropna(), bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Height Distribution')
axes[0].set_xlabel('Height (cm)')
axes[0].set_ylabel('Frequency')

# Distribution for weight
axes[1].hist(athletes['weight'].dropna(), bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title('Weight Distribution')
axes[1].set_xlabel('Weight (kg)')
axes[1].set_ylabel('Frequency')
```

### 4. Agrupe los datos por el calculo de la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

```python
grouped_by_sport = athletes.groupby('sport').mean(numeric_only=True)
deporte_mayor_peso = grouped_by_sport['weight'].idxmax()
peso_promedio_mayor = grouped_by_sport['weight'].max()

print(f"El deporte con el mayor peso promedio es {deporte_mayor_peso} con un peso promedio de {peso_promedio_mayor} kg.")
```

### 5. ¿Cuál es el pais con más participación total femenina?

```python
female_athletes = athletes[athletes['sex'] == 'female']

female_participation_by_country = female_athletes.groupby('nationality').size()


country_with_most_female_participation = female_participation_by_country.idxmax()
total_female_participation = female_participation_by_country.max()

print(f"El pais con la mayor participacion femenina es {country_with_most_female_participation} con {total_female_participation} participants.")
```
