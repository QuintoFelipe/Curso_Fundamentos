## **Solucion Ejercicio 1**

```python
# Importando librerias necesarias para el proceso
import pandas as pd
```

Creamos un DataFrame a partir del diccionario dado y usaremos las etiquetas especificadas como el indice    

```python
# Diccionario para la creacion del Dataframe con indices sugeridos en la actividad
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Creacion del Dataframe e impresion
df = pd.DataFrame(data=exam_data, index=labels)
print("DataFrame actividad:")
df
```

### 1. Seleccione solo las columnas *'name'* y *'score'*.
Para seleccionar columnas específicas, hacemos lo siguiente:

Seleccionando columnas con loc

```python
# Seleccionando columnas "name" y "score" con el accesor loc
print("Columnas seleccionadas 'name' y 'score':")
df.loc[:, ['name', 'score']]
```

Seleccionando columnas con iloc

```python
# Seleccionando columnas "name" y "score" con el accesor iloc
print("Columnas seleccionadas 'name' y 'score':")
# Opcion 1 indicando desde y hasta donde, no es incluyente en su finalizacion
df.iloc[:, 0:2]
# Opcion 2 en una lista se pasa las posiciones de las columnas requeridas
df.iloc[:, [0, 1]]
```

### 2. Filtre las filas donde el número de intentos (`attemps`) sea mayor que 2.

Podemos filtrar filas según una condición:

```python
# Creamos la mascara de booleanos de acuerdo a la condicion indicada
mask = df.attempts > 2
# Podemos identificar que las filas que cumplen con la condicion tienen valor True
mask
```

```python
# Aplicamos el filtro utilizando el accesor loc con nuestra variable mask
print("Filas donde attempts son mayores a 2:")
df.loc[mask]
```

### 3. Filtre las filas donde el *score* se encuentre entre 15 y 20 (incluidos)

Para filtrar basado en el puntaje, usamos:

Opción 1, creamos una mascara de booleanos basada en las condiciones indicadas

```python
# Creamos la variable mask que contiene el resultado booleano de la condicion expresada
mask = (df.score >= 15) & (df.score <= 20)
# Podemos identificar que las filas que cumplen con la condicion tienen valor True
mask
```

```python
# Usamos el accesor loc para realizar el filtrado basado en la mascara que obtuvimos
print("Filas donde el score está entre 15 y 20:")
df.loc[mask]
```

Opción 2, creamos una mascara de booleanos utilizando el metodo "between"

```python
# Creamos la variable mask que contiene el resultado booleano de la condicion expresada
mask = df.score.between(15, 20)
# Podemos identificar que las filas que cumplen con la condicion tienen valor True
mask
```

```python
# Usamos el accesor loc para realizar el filtrado basado en la mascara que obtuvimos
print("Filas donde el score está entre 15 y 20:")
df.loc[mask]
```

### 4. Cambie el *score* en la fila 'd' a 11.5

Podemos modificar el DataFrame utilizando los accesores loc e iloc.

```python
# Usamos el accesor loc para realizar el cambio indicado
print("DataFrame después de cambiar el score en la fila 'd' a 11.5:")
df.loc['d', 'score'] = 11.5
# Usamos el accesor iloc para realizar el cambio indicado
df.iloc[3, 1] = 11.5
df
```

### 5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

Podemos agregar una nueva fila utilizando el accesor `loc`:

```python
# Creamos la nueva fila con loc
print("DataFrame después de agregar la nueva fila 'k':")
df.loc['k'] = ['Nuevo Nombre', 18, 2, 'yes']
df
```

### 6. Borre la fila 'k' y entregue nuevamente el DataFrame original.

Para eliminar una fila, usamos el método `drop`es importante mencionar que podemos cambiar directamente el Dataframe usando la opción inplace en True, usamos la opción de errors="ignore" para aplicar sólo si existe el label y no indicar error en caso de que no sea así:



```python
print("DataFrame después de eliminar la fila 'k':")
# Realizamos el proceso creando una copia del resultado
df = df.drop('k', errors="ignore")
# Realizamos el proceso usando la opcion inplace en True
df.drop('k', inplace=True, errors="ignore")
df
```

### 7. Ordene el DataFrame por nombre en orden ascendente.

Podemos ordenar el DataFrame utilizando el método `sort_values`, es importante mencionar que podemos cambiar directamente el Dataframe usando la opción inplace en True:

```python
print("DataFrame ordenado por name:")
# Realizamos el proceso creando una copia del resultado
df = df.sort_values('name', ascending=True)
# Realizamos el proceso usando la opcion inplace en True
df.sort_values('name', ascending=True, inplace=True)
df
```

### 8. Reemplace los valores de la columna *qualify* que contiene los valores *yes* y *no* por los valores booleanos True y False respectivamente.

Finalmente, reemplazamos los valores en la columna qualify:

Podemos utilizar los métodos replace y map, son similares en cuanto a rendimiento y legibilidad, la ventaja de replace es que allí podemos incluir la opción inplace:

```python
# Creamos 3 copias del Dataframe para aplicar las diferentes opciones
df2 = df.copy()
df3 = df.copy()
df4 = df.copy()
```

```python
# Usamos el metodo map
print("DataFrame después de reemplazar 'yes' y 'no' en la columna qualify con True y False:")
df2.qualify = df.qualify.map({'yes': True, 'no': False})
df2
```

```python
# Usamos el metodo replace e incluimos la opcion inplace en True
print("DataFrame después de reemplazar 'yes' y 'no' en la columna qualify con True y False:")
df3.qualify.replace({'yes': True, 'no': False}, inplace=True)
df3
```

Podemos utilizar el método apply con lambda, es flexible pero suele ser menos eficiente que las otras opciones debido a que aplica la función a cada elemento individualmente:

```python
# Usamos el metodo apply con lambda
print("DataFrame después de reemplazar 'yes' y 'no' en la columna qualify con True y False:")
df4.qualify = df4.qualify.apply(lambda x: True if x == "yes" else False)
df4
```

Podemos utilizar el médotod numpy.where, puede ser más rápido en DataFrames grandes debido a su operación vectorizada:

```python
# Importamos numpy para realizar el proceso
import numpy as np
```

---

```python
# Usamos el metodo np.where
print("DataFrame después de reemplazar 'yes' y 'no' en la columna qualify con True y False:")
df.qualify = np.where(df.qualify == "yes", True, False)
df
```

## **Solucion Ejercicio 2**

### 1. Importarlo como nombre athletes

```python
# Importamos la libreria pandas
import pandas as pd

# Creamos el dataframe athletes segun lo indicado
print('DataFrame athletes importado:')
athletes = pd.read_csv('https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv')
athletes.head()
```

### 2.Filtrar el DataFrame para Quedarse Solo con los Atletas que Hayan Ganado al Menos una Medalla

Vamos a sumar las columnas de medallas (gold, silver, bronze) para determinar qué atletas han ganado al menos una medalla, este es un scope que vemos práctico según nuestras necesidades:

```python
# Creamos una nueva columna con la sumatoria de medallas
athletes['total_medals'] = athletes['gold'] + athletes['silver'] + athletes['bronze']
# Aplicamos el filtro con una mascara booleana
athletes = athletes.loc[athletes.total_medals > 0]
print("Atletas que han ganado al menos una medalla:")
athletes.head()
```

### 3. Calcule la masa muscular por fila con una función

Nos podemos dar cuenta que en las columnas de peso y altura hay valores nulos

```python
# Exploramos las diferentes columnas con el metodo info
athletes.info()
```

```python
# Importamos libreria necesaria para explorar graficamente
import matplotlib.pyplot as plt

# Plotting de las distribuciones
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Distribucion por altura
axes[0].hist(athletes['height'].dropna(), bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Height Distribution')
axes[0].set_xlabel('Height (cm)')
axes[0].set_ylabel('Frequency')

# Distribucion por peso
axes[1].hist(athletes['weight'].dropna(), bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title('Weight Distribution')
axes[1].set_xlabel('Weight (kg)')
axes[1].set_ylabel('Frequency')
```

Usamos imputacion mediante la media de los datos para el peso y la altura

```python
# Tranformamos las columnas weight y height modificando los valores nulos, y utilizamos la opcion inplace
athletes.fillna({'weight': athletes.weight.mean(), 'height': athletes.height.mean()}, inplace=True)
# Vemos los cambios
athletes.info()
```

Observamos la nueva distribución

```python
# Plotting las distribuciones
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Distribucion por altura
axes[0].hist(athletes['height'].dropna(), bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Height Distribution')
axes[0].set_xlabel('Height (cm)')
axes[0].set_ylabel('Frequency')

# Distribucion por peso
axes[1].hist(athletes['weight'].dropna(), bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title('Weight Distribution')
axes[1].set_xlabel('Weight (kg)')
axes[1].set_ylabel('Frequency')
```

Formúla para obtener el indice de masa muscular IMC = peso (kg) / altura^2 (m^2)

```python
# Se define la función para calcular la masa muscular
def calcular_masa_muscular(df):
    return df['weight'] / df['height']**2
```

```python
# Se crea la nueva columna 'masa muscular' utilizando el método apply y la función creada anteriormente
athletes['imm'] = athletes.apply(calcular_masa_muscular, axis=1)
athletes
```

### 4. Agrupe los datos calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

```python
# Usamos el metodo groupby para llegar al resultado solicitado
grouped_by_sport = athletes.groupby('sport').mean(numeric_only=True).sort_values('weight', ascending=False)
grouped_by_sport
```

```python
# Obtenemos el resultado y lo mostramos directamente desde nuestra tabla
print(f"El deporte con el mayor peso promedio es {grouped_by_sport.iloc[0].name} con un peso promedio de {round(grouped_by_sport.iloc[0].weight,2)} kg.")
```

### 5. ¿Cuál es el pais con más participación total femenina?

```python
# Usamos el metodo groupby con conteo para identificar el pais con mayor participacion femenina
female_participation_by_country = athletes.loc[:, ['nationality', 'sex']].loc[athletes.sex == 'female'].groupby('nationality').count().sort_values('sex', ascending=False)
female_participation_by_country
```

```python
# Obtenemos el resultado y lo mostramos directamente desde nuestra tabla
print(f"El pais con la mayor participacion femenina es {female_participation_by_country.iloc[0].name} con {female_participation_by_country.iloc[0].sex} participantes.")
```
